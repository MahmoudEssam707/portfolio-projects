### Import Libraries and Define Auxiliary Functions


We will import the following libraries the lab 



```python
# andas is a software library written for the Python programming language for data manipulation and analysis.
import pandas as pd
#NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays
import numpy as np
# Matplotlib is a plotting library for python and pyplot gives us a MatLab like plotting framework. We will use this in our plotter function to plot data.
import matplotlib.pyplot as plt
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns
```

```python
def Scatter(data,x, y, title, xlab, ylab):
    """Plotting Scatter with inputs to make it easier in printing"""
    plt.figure(figsize=(20,6))
    sns.scatterplot(x=x, y=y, hue="Class", data=data, palette='mako')
    plt.title(title, fontsize=30)
    plt.xlabel(xlab, fontsize=20)
    plt.ylabel(ylab, fontsize=20)
    hue = ['Fail', 'Success']
    handles, _ = plt.gca().get_legend_handles_labels()
    labels = hue
    plt.legend(handles, labels)
    plt.show();  
```

## Exploratory Data Analysis 


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary


```python
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")

df.head(5)
```

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass is also important; it seems the more massive the payload, the less likely the first stage will return.

We see that different launch sites have different success rates.  <code>CCAFS LC-40</code>, has a success rate of 60 %, while  <code>KSC LC-39A</code> and <code>VAFB SLC 4E</code> has a success rate of 77%.

# Patterns in Flight Number vs. Launch Site Scatter Point Plots

1. **CCAFS SLC 40 (Success Rate = 60%)**:
   - The scatter plot for CCAFS SLC 40 shows a relatively low success rate, at 60%.
   - There is a noticeable concentration of failed launches (points below the success rate line) compared to successful launches.
   - This launch site has seen a relatively high number of flight attempts.

2. **KSC LC 39A (Success Rate = 77.72%)**:
   - The scatter plot for KSC LC 39A displays a higher success rate, at 77.72%.
   - The majority of points on this plot are above the success rate line, indicating a higher proportion of successful launches.
   - KSC LC 39A has experienced a substantial number of flight attempts as well.

3. **VAFB SLC 4E (Success Rate = 76.9%)**:
   - VAFB SLC 4E's scatter plot shows a success rate of 76.9%, which is close to the success rate of KSC LC 39A.
   - There are relatively fewer data points compared to the other launch sites, suggesting that fewer launches have occurred here.
   - Despite the lower number of launches, there are relatively fewer failures, which may indicate a more reliable launch site.

**Interactions and Explanations:**

- While VAFB SLC 4E has had fewer launches compared to CCAFS SLC 40 and KSC LC 39A, it boasts a relatively high success rate. This could be attributed to a combination of factors, such as favorable launch conditions and rigorous quality control procedures.

- The fact that VAFB SLC 4E has experienced fewer failures, even with fewer launches, might be due to lessons learned from the higher number of launches at CCAFS SLC 40. In other words, the experience gained from the larger number of launch attempts at CCAFS SLC 40 may have contributed to better success rates at VAFB SLC 4E.

- It's noteworthy that some flights that failed at CCAFS SLC 40 were successful at other sites. This could be due to differences in launch conditions, equipment, or procedures at each site. It underscores the importance of considering multiple launch options to increase the chances of success.

In summary, the scatter point plots reveal variations in success rates and launch outcomes among the three launch sites. VAFB SLC 4E stands out with a relatively high success rate and fewer failures, potentially benefiting from the experience gained at CCAFS SLC 40, which has seen more launch attempts. These patterns highlight the complexity of rocket launches, where site-specific factors can significantly influence the success or failure of missions.

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column

plt.figure(figsize=(20, 6))
Orbit_VS_Class = df.groupby("Orbit")["Class"].mean() * 100  # Calculate the mean success percentage and convert to percentages
Orbit_VS_Class = Orbit_VS_Class.sort_values(ascending=False)

ax = sns.barplot(x=Orbit_VS_Class.values, y=Orbit_VS_Class.index, palette="mako")
plt.xlabel("Percentage of Success")
plt.ylabel("Orbit Type")
plt.title("Success Percentage by Orbit Type")

# Add percentage labels at the end of each bar
for index, value in enumerate(Orbit_VS_Class):
    plt.text(value, index, f"{value:.2f}%", va='center')

plt.show()
```

Analyze the ploted bar chart try to find which orbits have high sucess rate.

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend. 


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year(date):
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
years = Extract_year(df.Date)
df["Years"] = years
```

```python
Years_VS_Success = df.groupby("Years")["Class"].agg(np.mean)

# Creating a figure with improved style and setting the size
plt.figure(figsize=(20, 10))  # Adjust the DPI for higher resolution

# Creating the line plot using Seaborn with custom color and style
sns.lineplot(x=Years_VS_Success.index, y=Years_VS_Success.values * 100, color='royalblue', marker='o', linestyle='-', markersize=8, linewidth=2)

# Setting plot labels and title with improved font size and style
plt.xlabel("Years", fontsize=14, fontweight='bold')
plt.ylabel("Success (%)", fontsize=14, fontweight='bold')
plt.title("Years vs. Success", fontsize=16, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(["Success"], loc="upper left")
# Adding a background color to the plot
plt.gca().set_facecolor('whitesmoke')
# Adding data point annotations (optional)
for x, y in zip(Years_VS_Success.index, Years_VS_Success.values * 100):
    plt.annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0, 10), ha='center')
# Display the plot with tight layout
plt.tight_layout()
```

you can observe that the sucess rate since 2013 kept increasing till 2020
